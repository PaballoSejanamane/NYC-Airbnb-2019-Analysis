```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
airbnb = pd.read_csv("Airbnb_NYC_2019.csv")
airbnb
```

```python
#Getting Just Manhattan listings
manhattan = airbnb.loc[['Manhattan' in neighbourhood_group for neighbourhood_group in airbnb["neighbourhood_group"]], :]
manhattan
```

```python
#Number of rows
manhattan.shape[0]
```

```python
#Number of columns
manhattan.shape[1]
```

```python
#plotting a histogram for the prices for Airbnb listings in Manhattan
manhattan.hist(["price"], bins = np.arange(0, 2200, 100), color = 'orange')
plt.title("Distribution of Manhattan Airbnb listing Prices in 2019")
plt.xlabel("Price (in US Dollars)")
plt.ylabel("Count")
plt.show()
```

It seems that in 2019, the highest number of listings (approximately 8500) were made with price ranges of less thab $250 and the number of listing decreases to approximately 100 or less as the price for the Airbnb gets over $500. It seems the general populus is not willing to pay over $250 with 2100 Airbnbs going over $250

```python
plt.scatter(x= manhattan["price"], y= manhattan["number_of_reviews"], color ='violet')
plt.title("Scatter Plot of Number of Reviews vs Price in Manhattan in 2019")
plt.xlabel("Price (in US Dollars)")
plt.ylabel("Number of Reviews")
plt.show()
```

It seems that listings which were less than $1000 were more likely to leave reviews and listings more than $2000 had less than 50 reviews

```python
top_10 = airbnb.loc[['Manhattan' in neighbourhood_group for neighbourhood_group in airbnb["neighbourhood_group"]], :].sort_values("calculated_host_listings_count", ascending = False).iloc[:10]
top_10
```

```python
top_10['neighbourhood']
```

```python
top_10.shape[0]
```

```python
top_10.shape[1]
```

```python
sns.countplot(x = "neighbourhood" , data = top_10)
plt.title("Top 10 Neighbourhood Airbnb Listings in Manhattan in 2019")
plt.show()
```

Financial District and Murray Hill are the only two neighbourhoods in the top 10 and Financial District seems to have the most listings in Manhattan in 2019
